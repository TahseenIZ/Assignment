# Task: analyse the performance of an engineering students 8 semester CGPA with x axis as semester and y axis as CGPA

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
semesters = np.arange(1, 9)  
cgpa = np.array([3.5, 3.7, 3.8, 3.9, 3.6, 3.4, 3.2, 3.9])  # CGPA values
```

```python
plt.figure(figsize=(10, 6))
plt.plot(semesters, cgpa, color='red', linestyle='solid', marker="o", markerfacecolor='blue', markersize='12')
plt.title('Engineering Student CGPA Over 8 Semesters')
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.show()
```
