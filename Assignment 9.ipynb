```python
import csv
c=0
# opening the csv file
with open('C:/Users/amrit/Peace Dropbox/Amrit Kalmane/PC/Downloads/netflix_titles - netflix_titles.csv.csv',mode='r') as fd:
    #print("Successfully Opened csv")
    data = csv.reader(fd)
    print(type(data))
    for line in data:
        c = c+1
        if c>5:
            break
        if c==0:
            continue
        print(c)
        print('show_id:',line[0])
        print("Type:",line[1])
        print("Title:",line[2])
        print("Director:",line[3])
        print("Cast:",line[4])
        print("Country:",line[5])
        print()
```
