```python
import csv

# Create a sample dataset with male and female ratios
header = ['Group', 'Male Ratio', 'Female Ratio']
data = [
    ['Club', 60, 40],
    ['Event', 45, 55],
    ['Library ', 50, 50],
    ['Mall', 55, 45],
    ['Office', 48, 52],
]

# Save the dataset to a CSV file
csv_file_path = 'male_female.csv'
with open(csv_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(data)


import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset from the CSV file
data = pd.read_csv(csv_file_path)

# Pie Chart for Male Ratio
plt.figure(figsize=(8, 8))
plt.subplot(2, 3, 1)
plt.pie(data['Male Ratio'], labels=data['Group'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightskyblue'])
plt.title('Male Ratio in Different Groups')
plt.gcf().set_size_inches(8, 6)
plt.show()

# Pie Chart for Female Ratio
plt.subplot(2, 3, 3)
plt.pie(data['Female Ratio'], labels=data['Group'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightskyblue'])
plt.title('Female Ratio in Different Groups')

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

```python
# Column Chart for Male Ratio
plt.subplot(2, 1, 2)
plt.bar(data['Group'], data['Male Ratio'], color='lightblue')
plt.xlabel('Groups')
plt.ylabel('Male Ratio')
plt.title('Male Ratio in Different Groups (Column Chart)')

# Column Chart for Female Ratio
plt.figure(figsize=(4, 3))
plt.bar(data['Group'], data['Female Ratio'], color='pink')
plt.xlabel('Groups')
plt.ylabel('Female Ratio')
plt.title('Female Ratio in Different Groups (Column Chart)')
plt.show()

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

```python
# Histogram for Male Ratio
plt.subplot(2, 3, 2)
plt.hist(data['Male Ratio'], bins=10, color='lightblue', edgecolor='black')
plt.xlabel('Male Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Male Ratio')
plt.gcf().set_size_inches(8, 6)
plt.show()

# Histogram for Female Ratio
plt.figure(figsize=(6, 3))
plt.hist(data['Female Ratio'], bins=10, color='lightcoral', edgecolor='black')
plt.xlabel('Female Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Female Ratio')
plt.show()
# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

```python
# Scatter Chart for Male Ratio
plt.subplot(2, 2, 4)
plt.scatter(data['Group'], data['Male Ratio'], color='blue', label='Male Ratio')
plt.scatter(data['Group'], data['Female Ratio'], color='lightcoral', label='Female Ratio')
plt.xlabel('Groups')
plt.ylabel('Ratio')
plt.title('Male and Female Ratio in Different Groups (Scatter Chart)')
plt.legend()

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

```python
# Boxplot for Male Ratio
plt.subplot(4, 3, 2)
plt.boxplot(data['Male Ratio'])
plt.xticks([1], ['Male Ratio'])
plt.title('Boxplot of Male Ratio')
plt.gcf().set_size_inches(10, 8)
plt.show()

# Boxplot for Female Ratio
plt.figure(figsize=(4, 4))
plt.boxplot(data['Female Ratio'], vert=False)
plt.yticks([1], ['Female Ratio'])
plt.title('Boxplot of Female Ratio')
plt.show()

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```
