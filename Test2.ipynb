```python
import numpy as np
dataset=np.array([2,4,5,2,8,4,3,7,8,2])
mean=np.mean(dataset)
median=np.median(dataset)
mode=np.argmax(np.bincount(dataset))
print(f"mean:{mean},median:{median},mode:{mode}")
```

```python
#partc
import numpy as np
exam_scores = np.array([85, 90, 88, 78, 92, 95, 82, 88, 90, 85])
mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)
std_deviation = np.std(exam_scores)
variance = np.var(exam_scores)

print(f"Mean: {mean_score}, Median: {median_score}")
print(f"Standard Deviation: {std_deviation}, Variance: {variance}")
```

```python
import matplotlib.pyplot as plt
import numpy as np


experience, salary = np.random.randint(1, 20, size=50), 3000 + 500 * np.random.randint(1, 20, size=50) + np.random.normal(0, 2000, size=50)
plt.scatter(experience, salary, alpha=0.7)

plt.title('Experience vs Salary')
plt.xlabel(' Experience (years)')
plt.ylabel('Salary ')

plt.grid(True)
plt.show()
```

```python

def cal_bmi(weight, height):
    return weight / (height ** 2)

def health_status(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'
weight = 70
height = 1.75
bmi = cal_bmi(weight, height)
print(f'BMI: {bmi:.2f}')
print(f'Health Status: {health_status(bmi)}')
```
